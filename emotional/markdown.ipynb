# markdown常用编写方式
**Markdown是一种轻量级标记语言，使用纯文本格式编写文档，然后转换为HTML。它非常适合编写文档、README文件、博客等。以下是一些常用的Markdown编写方式：**

## 标题
**使用井号（#）表示标题，井号的数量决定标题的级别：**

*举例：*
# 一级标题
## 二级标题
### 三级标题

## 强调
**使用星号（*）或下划线（_），表示强调:**

*一个星或下划线举例：*
*斜体*

*两个星/下滑线举例：*
__粗体__

*三个星/下划线举例：*
***加粗斜体***

## 列表
**无序列表使用星号（*）、加号（+）或减号（-）表示：**

- 项目一
- 项目二
  - 子项目
  - 子项目

+ 项目一
+ 项目二

* 项目一
* 项目二

**有序列表使用数字和点号：**

1. 项目一
2. 项目二
   1. 子项目一
   2. 子项目二

## 链接
**使用方括号[文本]和圆括号(链接地址)添加链接：**

[链接文本](https://baidu.com)

## 图片
**与链接类似，但前面加一个感叹号：**

![替代文本](https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcSW6803GR4h7m8SMuyDz0Yo38lVgAjeNzZHcA&s)

## 引用
**使用大于号（>）表示引用：**

> 这是一个引用。

## 代码
**单行代码使用反引号（`）：**

这是`代码`。
多行代码使用三个反引号：
这是
```多行代码```。

## 表格
**使用竖线和横线创建表格：**

| 标题1 | 标题2 |
| ----- | ----- |
| 内容1 | 内容2 |
| 内容3 | 内容4 |
### 分隔线
**使用三个以上的星号、短横线或下划线表示分隔线：**

***
